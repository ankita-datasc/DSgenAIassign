```python
#univariate analysis
#categorical columns
cat_cols=df[['GENDER', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
       'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
       'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']]

for col in cat_cols:
    plt.figure(figsize=(10,5))
    plt.title(f"Count plot of {col}")
    sns.countplot(data=df,x=col,palette='viridis')
    plt.xlabel("1:NO and 2:Yes")
    plt.xticks(rotation=45)
    plt.show()

print("\n")
#Age is a numerical column
plt.figure(figsize=(10,5))
plt.title(f"Distribution plot of Age")
sns.histplot(data=df,x=df.AGE,kde=True,palette='viridis')
plt.show()
```

```python
#label encoding Gender and lung cancer column to turn them to numeric columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['GENDER']=le.fit_transform(df['GENDER'])
df['LUNG_CANCER']=le.fit_transform(df['LUNG_CANCER'])
```
