```python
cat_cols=df[['sex', 'federal_district', 'type_of_city', 'knows_election_date',
       'will_vote', 'television_usage', 'internet_usage',
       'education', 'income', 'employment', 'job_type']]
for col in cat_cols:
  plt.figure(figsize=(10,4))
  sns.countplot(data=df, x=col, palette='viridis')
  plt.title(f'Count Plot of {col}')
  plt.xticks(rotation=45)
  plt.show()
```

```python
cat_cols=df[['sex', 'federal_district', 'type_of_city', 'knows_election_date',
       'will_vote', 'television_usage', 'internet_usage',
       'education', 'income', 'employment', 'job_type']]
for col in cat_cols:
  plt.figure(figsize=(12,4))
  sns.countplot(data=df, x=col, hue='candidate')
  plt.title(f'Count Plot of {col}')
  plt.xticks(rotation=45)
  plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='candidate', y='age',palette='colorblind')
plt.title('Box Plot of Age by Candidate')
plt.xlabel('Candidate')
plt.xticks(rotation=45)
plt.ylabel('Age')
plt.show()
```

```python
for col in cat_cols:
  plt.figure(figsize=(6,6))
  plt.pie(df[col].value_counts(), labels=df[col].value_counts().index, autopct='%1.1f%%',explode=[0.05]*len(df[col].value_counts()),shadow=True)
  plt.title(f'Pie Chart of {col}')
  plt.show()
  print("\n")
```
